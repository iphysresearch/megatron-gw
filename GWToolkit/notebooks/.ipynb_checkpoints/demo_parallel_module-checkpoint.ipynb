````python
def time_waveform_response_block(self, num, start_time, geocent_time=None, target_optimal_snr_tuple=None,
                                 disable=True):
    """Generate a data block with detector responses of GW waveform in time domain
    num: int
        Number of data
    geocent_time: tuple, optional, (minimum, maximum)
        The GPS arrival time of the signal data
    start_time: float (default: 0)
        The GPS start-time of the data / Starting time of the time array
    target_optimal_snr_tuple： tuple, optional, (target_detector_index, target_optimal_snr)

    #TODO too-many-locals
    """
    Record = namedtuple('Record', 'block noise optimal_snr matched_filter_snr meta_data block_whitened signal_whitened noise_whitened')
    block = Record(numpy.empty(shape=(0, len(self.dets.keys()), self.num_t), dtype=numpy.float64),
                   numpy.empty(shape=(0, len(self.dets.keys()), self.num_t), dtype=numpy.float64),
                   numpy.empty(shape=(0, len(self.dets.keys())), dtype=numpy.float64),
                   numpy.empty(shape=(0, len(self.dets.keys())), dtype=numpy.complex64),
                   {},
                   numpy.empty(shape=(0, len(self.dets.keys()), self.num_t), dtype=numpy.float64),
                   numpy.empty(shape=(0, len(self.dets.keys()), self.num_t), dtype=numpy.float64),
                   numpy.empty(shape=(0, len(self.dets.keys()), self.num_t), dtype=numpy.float64))
    # meta_data = {}
    alpha = 1
    # noise_freq_block = self.frequency_colored_noise_block(num)
    # noise_block = numpy.empty(shape=(num, len(self.dets.keys()), self.num_t), dtype=numpy.float64)
    # optimal_snr = numpy.empty(shape=(num, len(self.dets.keys())), dtype=numpy.float64)
    # matched_filter_snr = numpy.empty(shape=(num, len(self.dets.keys())), dtype=numpy.complex64)
    # block = numpy.empty(shape=(num, len(self.dets.keys()), self.num_t), dtype=numpy.float64)
    # snr_flat = numpy.random.rand(num) * 45 + 5
    @ray.remote
    def _inside_func():
        self._update_waveform()
        self.update_detector_response(start_time=start_time, geocent_time=geocent_time)

        if isinstance(target_optimal_snr_tuple, Iterable) and target_optimal_snr_tuple[1]:

            optimal_snr_cache = self.dets[list(self.dets.keys())[target_optimal_snr_tuple[0]]].optimal_snr(
                self.frequency_waveform_response[target_optimal_snr_tuple[0]]
            )
            alpha = target_optimal_snr_tuple[1] / optimal_snr_cache
            optimal_snr = numpy.concatenate(
                [numpy.asarray([target_optimal_snr_tuple[1]])
                 if j == target_optimal_snr_tuple[0] else
                 numpy.asarray([det.optimal_snr(self.frequency_waveform_response[j] * alpha)])
                 for j, (_, det) in enumerate(self.dets.items())
                 ],
            )
            # block.optimal_snr[i] = snr_flat[i]
            # alpha = snr_flat[i] / optimal_snr_cache
        else:
            optimal_snr = numpy.concatenate(
                [numpy.asarray([det.optimal_snr(
                 self.frequency_waveform_response[j] * alpha)])
                 for j, (_, det) in enumerate(self.dets.items())
                 ],
            )

        block = self.time_waveform_response * alpha
        block_whitened = numpy.concatenate(
            [numpy.asarray([det.frequency_to_time_domain(det.whiten(
             self.frequency_waveform_response[j] * alpha + self.frequency_colored_noise[j]))[0]])
             for j, (_, det) in enumerate(self.dets.items())
             ],
        )

        signal_whitened = numpy.concatenate(
            [numpy.asarray([det.frequency_to_time_domain(det.whiten(
             self.frequency_waveform_response[j] * alpha ))[0]])
             for j, (_, det) in enumerate(self.dets.items())
             ],
        )

        noise_whitened = numpy.concatenate(
            [numpy.asarray([det.frequency_to_time_domain(det.whiten(
             self.frequency_colored_noise[j]))[0]])
             for j, (_, det) in enumerate(self.dets.items())
             ],
        )

        matched_filter_snr = numpy.concatenate(
            [numpy.asarray([det.matched_filter_snr(
             self.frequency_waveform_response[j] * alpha,
             self.frequency_waveform_response[j] * alpha + self.frequency_colored_noise[j])])
             for j, (_, det) in enumerate(self.dets.items())
             ],
        )
        
        # meta_data = {}
        # for key, value in self.parameters.items():
        #     if key not in meta_data:
        #         meta_data[key] = numpy.empty([num, ])
        #     meta_data[key][i] = value

        noise = numpy.concatenate(
            [det.frequency_to_time_domain(self.frequency_colored_noise[j])[0][numpy.newaxis,...]
            for j, (_, det) in enumerate(self.dets.items())
            ],
        )

        return (block[numpy.newaxis,...], noise[numpy.newaxis,...],
                block_whitened[numpy.newaxis,...], 
                signal_whitened[numpy.newaxis,...], noise_whitened[numpy.newaxis,...], 
                optimal_snr, matched_filter_snr, 
                self.parameters)

    # Way 1
    data = ray.get([_inside_func.remote() for _ in tqdm(range(num))])
    data = numpy.asarray(data, dtype=object)
    meta_data = {key: numpy.concatenate([data[i,-1][key] for i in range(num)]) for key in data[0,-1].keys()}
    meta_data['optimal_snr'] = numpy.concatenate(data[:,5])
    meta_data['matched_filter_snr'] = numpy.concatenate(data[:,6])
    return (numpy.concatenate(data[:,0]),
        meta_data,
        numpy.concatenate(data[:,1]),
        numpy.concatenate(data[:,2]),
        numpy.concatenate(data[:,3]),
        numpy.concatenate(data[:,4]))

    # Way 2
    def process_incremental(data, new_data):
        # new_data = numpy.asarray(new_data, dtype=object)
        data[0] = numpy.concatenate([data[0], new_data[0]])
        data[1] = numpy.concatenate([data[1], new_data[1]])
        data[2] = numpy.concatenate([data[2], new_data[2]])
        data[3] = numpy.concatenate([data[3], new_data[3]])
        data[4] = numpy.concatenate([data[4], new_data[4]])
        return data

    data_ids = [_inside_func.remote() for _ in tqdm(range(num))]
    
    # data = [block.block, block.noise, block.block_whitened, block.signal_whitened, block.noise_whitened, block.meta_data]
    block = numpy.empty(shape=(0, len(self.dets.keys()), self.num_t), dtype=numpy.float64)
    noise = numpy.empty(shape=(0, len(self.dets.keys()), self.num_t), dtype=numpy.float64)
    block_whitened = numpy.empty(shape=(0, len(self.dets.keys()), self.num_t), dtype=numpy.float64)
    signal_whitened = numpy.empty(shape=(0, len(self.dets.keys()), self.num_t), dtype=numpy.float64)
    noise_whitened = numpy.empty(shape=(0, len(self.dets.keys()), self.num_t), dtype=numpy.float64)
    while len(data_ids):
        done_id, data_ids = ray.wait(data_ids)
        # data = process_incremental(data, ray.get(done_id[0]))
        # data = ray.get(done_id[0])
        block = numpy.concatenate([block, ray.get(done_id[0])[0]])
        noise = numpy.concatenate([noise, ray.get(done_id[0])[1]])
        block_whitened = numpy.concatenate([block_whitened, ray.get(done_id[0])[2]])
        signal_whitened = numpy.concatenate([signal_whitened, ray.get(done_id[0])[3]])
        noise_whitened = numpy.concatenate([noise_whitened, ray.get(done_id[0])[4]])
    return block, noise, block_whitened, signal_whitened, noise_whitened

#     block.block, block.meta_data, block.noise, block.block_whitened, block.signal_whitened, block.noise_whitened
    
#     meta_data = {key: numpy.concatenate([data[i,-1][key] for i in range(num)]) for key in data[0,-1].keys()}
#     meta_data['optimal_snr'] = numpy.concatenate(data[:,5])
#     meta_data['matched_filter_snr'] = numpy.concatenate(data[:,6])
#     return (numpy.concatenate(data[:,0]),
#     meta_data,
#     numpy.concatenate(data[:,1]),
#     numpy.concatenate(data[:,2]),
#     numpy.concatenate(data[:,3]),
#     numpy.concatenate(data[:,4]))
````

````python
from numpy.fft import rfft, irfft, rfftfreq
from scipy.interpolate import interp1d
from gwpy.timeseries import TimeSeries
from gwpy.signal import filter_design


def whiten(signal, sampling_rate, f_ASD, ASD, return_tilde=False):

    f_signal = rfftfreq(len(signal), 1./sampling_rate)
    psd = np.abs(interp1d(f_ASD, ASD**2, bounds_error=False, fill_value=np.inf)(f_signal))
    signal_filtered_tilde = rfft(signal) / np.sqrt(0.5 * sampling_rate * psd)
    if return_tilde:
        return irfft(signal_filtered_tilde), signal_filtered_tilde
    else:
        return irfft(signal_filtered_tilde)
````

## Whitening

````python
from numpy.fft import rfft, irfft, rfftfreq
from scipy.interpolate import interp1d
from scipy.signal import welch

def whiten(signal, sampling_rate, f_ASD=None, ASD=None, return_tilde=False, return_psd=False):
    """https://github.com/moble/MatchedFiltering/blob/3d7f3b5ca1383e26b80e64c82a703fe054f16c40/utilities.py#L155
    """
    f_signal = rfftfreq(len(signal), 1./sampling_rate)
    if ASD is not None:
        psd = np.abs(interp1d(f_ASD, ASD**2, bounds_error=False, fill_value=np.inf)(f_signal))
    else:
        f_psd, psd = welch(signal, sampling_rate, nperseg=2**int(np.log2(len(signal)/8.0)), scaling='density')
        psd = np.abs(interp1d(f_psd, psd, bounds_error=False, fill_value=np.inf)(f_signal))
    signal_filtered_tilde = rfft(signal) / np.sqrt(0.5 * sampling_rate * psd)
    if return_tilde:
        return irfft(signal_filtered_tilde), signal_filtered_tilde
    elif return_psd:
        return irfft(signal_filtered_tilde), (f_signal, psd)
    else:
        return irfft(signal_filtered_tilde)
````

## Standardize for strains

````python
"""
Transforms
"""
import sys
import numpy

class Normalize_strain():
    """Standardize for strains
    """
    def __init__(self, feature_range=(-1, 1), verbose=True, **kwargs):
        # Print debug messages or not?
        self.verbose = verbose
        self.feat_min, self.feat_max = feature_range

    def transform(self, X, Xmin=None, Xmax=None):
        """'minimax' method to standardizing
        Input
            X: ndarray
            Xmin:
            Xmax:

        Return
            X_scaled: same shape of X
            Xmin[...,0,0]: array with length of X
            Xmax[...,0,0]: array with length of X
        """
        if Xmin is None:
            Xmin = X.min(axis=-1, keepdims=True).min(axis=-2, keepdims=True)
        if Xmax is None:
            Xmax = X.max(axis=-1, keepdims=True).max(axis=-2, keepdims=True)
        X_std = (X - Xmin) / (Xmax - Xmin)
        X_scaled = X_std * (self.feat_max - self.feat_min) + self.feat_min
        return X_scaled, Xmin, Xmax

    def inverse_transform(self, X_scaled, Xmin, Xmax):
        """'minimax' method to inverse standardizing
        Input:
            X_scaled: ndarray
            maxx: array with length of X
            minn: array with length of X
        Return
            :same shape of X
        """
        # Xmax = maxx[...,np.newaxis,np.newaxis]
        # Xmin = minn[...,np.newaxis,np.newaxis]
        return (X_scaled - self.feat_min) * (Xmax - Xmin) / (self.feat_max - self.feat_min) + Xmin

    def vprint(self, string, *args, **kwargs):
        """
        Verbose printing: Wrapper around `print()` to only call it if
        `self.verbose` is set to true.

        Args:
            string (str): String to be printed if `self.verbose`
                is `True`.
            *args: Arguments passed to `print()`.
            **kwargs: Keyword arguments passed to `print()`.
        """

        if self.verbose:
            print(string, *args, **kwargs)
            sys.stdout.flush()
````

````python
def calc_matches(d1, d2):
    fft1 = np.fft.fft(d1)
    fft2 = np.fft.fft(d2)
    norm1 = np.mean(np.abs(fft1)**2)
    norm2 = np.mean(np.abs(fft2)**2)
    inner = np.mean(fft1.conj()*fft2).real
    return inner / np.sqrt(norm1 * norm2)
````

````python
# 1. Init for WaveformDataset
sampling_frequency = 4096     # [Hz], sampling rate
duration_long = 32                  # [sec], duration of a sample
duration = 8                  # [sec], duration of a sample
conversion = 'BBH'
waveform_approximant = 'IMRPhenomPv2'
# waveform_approximant = 'SEOBNRv4P' ## TODO
# waveform_approximant = 'IMRPhenomPv2_NRTidal'
# waveform_approximant = 'IMRPhenomXP'
reference_frequency = 50.
minimum_frequency = 20.
waveform_arguments = dict(waveform_approximant=waveform_approximant,
                          reference_frequency=reference_frequency,
                          minimum_frequency=minimum_frequency)
base = 'bilby'
dets = ['H1', 'L1'][:1]

# filename = '/workspace/zhaoty/GWToolkit/tests/gw/demo.prior'   # default prior file
filename = '/workspace/zhaoty/GWToolkit/gwtoolkit/gw/prior_files/default.prior'   # default prior file

# waveform dataset
wfd = WaveformDataset(sampling_frequency=sampling_frequency,
                      duration=duration_long,
                      conversion=conversion)

wfd.load_prior_source_detector(
    filename=filename,
    base=base,
    dets=dets,
    waveform_arguments=waveform_arguments)


####################################################################################
# Set for waveforms
# target_time = 1126259456.3999023 + 6
target_time = 1126259462.425#GWTC1_events['GW150914']['trigger-time'] 
buffer_time = duration_long / 2 - duration /4
start_time = target_time-(duration_long - buffer_time)
geocent_time = (target_time-1.1, target_time+1.1)


####################################################################################
# Set for noises
data_dir = '/workspace/zhaoty/dataset/O1_H1_All/'
noise_interval = 1024
selected_hdf_file_ratio = 0.5
num_length = 2
verbose = False

def cut_from_long(time, data):
    left_index = int((duration_long - duration)/2*sampling_frequency)
    right_index = int((duration_long + duration)/2*sampling_frequency)
    return time[left_index: right_index], data[left_index: right_index]


def whiten(signal, sampling_rate, f_ASD=None, ASD=None, return_tilde=False, return_psd=False):
    """https://github.com/moble/MatchedFiltering/blob/3d7f3b5ca1383e26b80e64c82a703fe054f16c40/utilities.py#L155
    """
    f_signal = rfftfreq(len(signal), 1./sampling_rate)
    if ASD is not None:
        psd = np.abs(interp1d(f_ASD, ASD**2, bounds_error=False, fill_value=np.inf)(f_signal))
    else:
        f_psd, psd = welch(signal, sampling_rate, nperseg=2**int(np.log2(len(signal)/8.0)), scaling='density')
        psd = np.abs(interp1d(f_psd, psd, bounds_error=False, fill_value=np.inf)(f_signal))
    signal_filtered_tilde = rfft(signal) / np.sqrt(0.5 * sampling_rate * psd)
    if return_tilde:
        return irfft(signal_filtered_tilde), signal_filtered_tilde
    elif return_psd:
        return irfft(signal_filtered_tilde), (f_signal, psd)
    else:
        return irfft(signal_filtered_tilde)

addr_asds = [
    # https://dcc.ligo.org/T1800044-v5/public
    '/workspace/zhaoty/GWToolkit/gwtoolkit/gw/prior_files/aLIGODesign.txt', 
    # https://dcc.ligo.org/LIGO-T2000012/public
    '/opt/conda/lib/python3.8/site-packages/bilby/gw/detector/noise_curves/aLIGO_ZERO_DET_high_P_asd.txt',
    '/opt/conda/lib/python3.8/site-packages/bilby/gw/detector/noise_curves/aLIGO_early_asd.txt',
    '/opt/conda/lib/python3.8/site-packages/bilby/gw/detector/noise_curves/aLIGO_early_high_asd.txt',
    '/opt/conda/lib/python3.8/site-packages/bilby/gw/detector/noise_curves/aLIGO_mid_asd.txt',
    '/opt/conda/lib/python3.8/site-packages/bilby/gw/detector/noise_curves/aLIGO_late_asd.txt',
]
````

````python
import sys
sys.path.append('..')
from gwtoolkit.gw import WaveformDataset
from gwtoolkit.torch import (WaveformDatasetTorch, Normalize_params, Patching_data, ToTensor)
import torch
from torchvision import transforms
from torch.utils.data import DataLoader
import itertools
import numpy as np

import matplotlib.pyplot as plt
from tqdm import tqdm
import time

from numpy.fft import rfft, irfft, rfftfreq
from scipy.interpolate import interp1d
from scipy.signal import welch

from bilby.gw import utils as gwutils

class RayDatasetTorch():
    """
    Dataset for Ray and Torch
    """
    def __init__(self):
        """Instantiate a ...
        """
        # 1. Init for WaveformDataset
        self.sampling_frequency = 4096     # [Hz], sampling rate
        self.duration_long = 32            # [sec], duration of a sample
        self.duration = 8                  # [sec], duration of a sample
        conversion = 'BBH'
        waveform_approximant = 'IMRPhenomPv2'
        # waveform_approximant = 'SEOBNRv4P' ## TODO
        # waveform_approximant = 'IMRPhenomPv2_NRTidal'
        # waveform_approximant = 'IMRPhenomXP'
        reference_frequency = 50.
        minimum_frequency = 20.
        waveform_arguments = dict(waveform_approximant=waveform_approximant,
                                  reference_frequency=reference_frequency,
                                  minimum_frequency=minimum_frequency)
        base = 'bilby'
        self.dets = ['H1', 'L1'][:1]  # TODO 

        # filename = '/workspace/zhaoty/GWToolkit/tests/gw/demo.prior'   # default prior file
        filename = '/workspace/zhaoty/GWToolkit/gwtoolkit/gw/prior_files/default.prior'   # default prior file

        # waveform dataset
        self.wfd = WaveformDataset(sampling_frequency=self.sampling_frequency,
                                   duration=self.duration_long,
                                   conversion=conversion)

        self.wfd.load_prior_source_detector(
            filename=filename,
            base=base,
            dets=self.dets,
            waveform_arguments=waveform_arguments)

        ####################################################################################
        # Set for waveforms
        # target_time = 1126259456.3999023 + 6
        self.target_time = 1126259462.425#GWTC1_events['GW150914']['trigger-time'] 
        buffer_time = self.duration_long / 2 - self.duration /4
        self.start_time = self.target_time-(self.duration_long - buffer_time)
        self.geocent_time = (self.target_time-1.1, self.target_time+1.1)


        ####################################################################################
        # Set for noises
        self.data_dir = '/workspace/zhaoty/dataset/O1_H1_All/'
        self.noise_interval = 1024
        self.selected_hdf_file_ratio = 0.5
        self.num_length = 2

        # Used for estimate PSDs
        self.seg_sec = 1
        self.random_timedomain_ratio = 0.3

        # Used for generate backgrounds
        self.freq = None
        self.Pxx = None

        self.addr_asds = [
            # https://dcc.ligo.org/T1800044-v5/public
            '/workspace/zhaoty/GWToolkit/gwtoolkit/gw/prior_files/aLIGODesign.txt', 
            # https://dcc.ligo.org/LIGO-T2000012/public
            '/opt/conda/lib/python3.8/site-packages/bilby/gw/detector/noise_curves/aLIGO_ZERO_DET_high_P_asd.txt',
            '/opt/conda/lib/python3.8/site-packages/bilby/gw/detector/noise_curves/aLIGO_early_asd.txt',
            '/opt/conda/lib/python3.8/site-packages/bilby/gw/detector/noise_curves/aLIGO_early_high_asd.txt',
            '/opt/conda/lib/python3.8/site-packages/bilby/gw/detector/noise_curves/aLIGO_mid_asd.txt',
            '/opt/conda/lib/python3.8/site-packages/bilby/gw/detector/noise_curves/aLIGO_late_asd.txt',
        ]
        
        # Set for patching tokens
        patch_size = 0.5  # [sec]
        overlap = 0.5     # [%]
        self.patching = Patching_data(
            patch_size=patch_size,
            overlap=overlap,
            sampling_frequency=self.sampling_frequency
        )
        # Init
        self.update_waveforms()
        self.update_backgrounds(4)

    def update_backgrounds(self, level):
        """
        Level:
        4: Reselect % of hdf5s -> Resample strains from selected hdf5s -> Restimate PSD by resegmenting -> Regenete BG from PSD
        3:                        Resample strains from selected hdf5s -> Restimate PSD by resegmenting -> Regenete BG from PSD
        2:                                                                Restimate PSD by resegmenting -> Regenete BG from PSD
        1:                                                                                                 Regenete BG from PSD        
        """
        level_list = [0]*(4-level)+[1]*level
        # Loop for noises
        for det in self.dets:
            if level_list[0]:
                # Select % of hdf5 files (Randomly)
                self.wfd.dets[det].load_from_GWOSC(self.data_dir, self.noise_interval, self.selected_hdf_file_ratio,
                                                   self.num_length, verbose=False)
        
            if level_list[1]:
                # Sampling strain_multi and time_multi (Randomly)
                self.wfd.dets[det].gwosc.update_randomly_strain()
            if level_list[2]:
                # Estimate PSD from strain_multi/time_multi using welch_modified (Randomly)
                self.freq, self.Pxx = self.wfd.dets[det].gwosc.estimate_randomly_psd_cvmfs(self.seg_sec, 
                                                                                 self.random_timedomain_ratio,
                                                                                 noverlap=0)
            if level_list[3]:
                # It will cover the PSD and generate new noises (Randomly)
                self.wfd.dets[det].ifo.power_spectral_density = \
                self.wfd.dets[det].ifo.power_spectral_density.from_power_spectral_density_array(self.freq, self.Pxx)
                self.wfd.dets[det].update() # corresponding to duration_long
                # wfd.dets[det].ifo.maximum_frequency = 512  # Can it be used?

    def update_waveforms(self):
        # Loop for waveforms
        self.wfd._update_waveform()  # update internal parameters

        # update external parameters / start_time
        self.wfd.update_detector_response(start_time=self.start_time, geocent_time=self.geocent_time)

    def generate_sample(self, use_which_design_ASD_for_whiten=None, target_optimal_snr=None):

        alpha = 1
        # TODO need to support multi-detectors
        det = 'H1'
        if use_which_design_ASD_for_whiten is not None:
            ## designed PSDs
            data = np.loadtxt(addr_asds[2])
            ASDf, ASD = data[:,0], data[:,1]
        else:
            ## current on-source PSDs
            ASDf, ASD = None, None

        # whiten the background
        noise, (fpsd, psd) = self.whiten(self.wfd.dets[det].ifo.strain_data.time_domain_strain, 
                                         self.sampling_frequency, ASDf, ASD, return_psd=True)
        signal = self.whiten(self.wfd.time_waveform_response[0], # H1 # TODO
                             self.sampling_frequency, fpsd, np.sqrt(psd))

        if target_optimal_snr is not None:
            optimal_snr, _ = self.calc_SNR(alpha, psd)
            alpha = target_optimal_snr / optimal_snr
            optimal_snr_, matched_filter_snr = self.calc_SNR(alpha, psd)
            assert np.allclose(target_optimal_snr,optimal_snr_), f'target_optimal_snr:{target_optimal_snr} vs optimal_snr_:{optimal_snr_}'
            data = signal * alpha + noise
            return data, signal, noise, target_optimal_snr, matched_filter_snr            
        else:
            optimal_snr, matched_filter_snr = self.calc_SNR(alpha, psd)

            data = signal * alpha + noise
            return data, signal, noise, optimal_snr, matched_filter_snr


    def calc_SNR(self, alpha, psd):
        # TODO need to support multi-detectors
        det = 'H1'        
        data = self.wfd.dets[det].ifo.strain_data.time_domain_strain + self.wfd.time_waveform_response[0] * alpha
        return \
        np.sqrt(gwutils.optimal_snr_squared(
            signal=self.wfd.frequency_waveform_response[0][self.wfd.dets[det].ifo.strain_data.frequency_mask] * alpha,
            power_spectral_density=psd[self.wfd.dets[det].ifo.strain_data.frequency_mask],
            duration=self.wfd.dets[det].ifo.strain_data.duration).real) ,\
        gwutils.matched_filter_snr(
                    signal=self.wfd.frequency_waveform_response[0][self.wfd.dets[det].ifo.strain_data.frequency_mask] * alpha,
                    frequency_domain_strain=self.wfd.dets[det].time_to_frequency_domain(data)[0][self.wfd.dets[det].ifo.strain_data.frequency_mask],
                    power_spectral_density=psd[self.wfd.dets[det].ifo.strain_data.frequency_mask],
                    duration=self.wfd.dets[det].ifo.strain_data.duration)

        
    def cut_from_long(self,time, data):
        left_index = int((self.duration_long - self.duration)/2*self.sampling_frequency)
        right_index = int((self.duration_long + self.duration)/2*self.sampling_frequency)
        return time[left_index: right_index], data[left_index: right_index]


    def whiten(self, signal, sampling_rate, f_ASD=None, ASD=None, return_tilde=False, return_psd=False):
        """
        Ref: 
            https://github.com/moble/MatchedFiltering/blob/3d7f3b5ca1383e26b80e64c82a703fe054f16c40/utilities.py#L155
        """
        f_signal = rfftfreq(len(signal), 1./sampling_rate)
        if ASD is not None:
            psd = np.abs(interp1d(f_ASD, ASD**2, bounds_error=False, fill_value=np.inf)(f_signal))
        else:
            f_psd, psd = welch(signal, sampling_rate, nperseg=2**int(np.log2(len(signal)/8.0)), scaling='density')
            psd = np.abs(interp1d(f_psd, psd, bounds_error=False, fill_value=np.inf)(f_signal))
        signal_filtered_tilde = rfft(signal) / np.sqrt(0.5 * sampling_rate * psd)
        if return_tilde:
            return irfft(signal_filtered_tilde), signal_filtered_tilde
        elif return_psd:
            return irfft(signal_filtered_tilde), (f_signal, psd)
        else:
            return irfft(signal_filtered_tilde)

````

````python
def toRedis(value,name):
    """Store given Numpy array 'value' in Redis under key 'name'"""
    r.set(name,m.packb(value))
    return

def fromRedis(r,n):
    """Retrieve Numpy array from Redis key 'n'"""
    # Retrieve and unpack the data
    try:
        return m.unpackb(r.get(n))
    except TypeError:
        print('No this value')

def set_get_Redis(r, value, name):
    try:
        return m.unpackb(r.set(name, m.packb(value), get=True))    
    except TypeError:
        return None
````

````python
srcFile = open('loss.log', 'r+')
lines = srcFile.readlines()
loss = []
for line in tqdm(lines):
    d = eval(line.split('|')[5].split(':')[1])
    loss.append(d)
srcFile.close()

srcFile = open('match.log', 'r+')
lines = srcFile.readlines()
matches = []
for line in tqdm(lines):
    d1, d2 = eval(line.split('|')[0].split('match long value: ')[-1]), eval(line.split('|')[1].split('match short value:')[-1])
    matches.append([d1, d2])
matches = np.asarray(matches)
srcFile.close()

srcFile = open('validation_loss.log', 'r+')
lines = srcFile.readlines()
loss_valid = []
for line in tqdm(lines):
    d = eval(line.split('|')[1].split('lm loss value:')[-1])
    loss_valid.append(d)
srcFile.close()
````

````python
plt.semilogy(loss, 'k-', label='Loss')
plt.plot(np.linspace(0, stop=len(loss), num=len(loss_valid)), loss_valid, 'k:', label='Valid Loss')
plt.legend();
plt.ylabel('Loss')
plt.twinx()
plt.plot(np.linspace(0, stop=len(loss), num=len(matches)), matches[:,0], label='Valid on Long')
plt.plot(np.linspace(0, stop=len(loss), num=len(matches)), matches[:,1], label='Valid on Short')
plt.ylabel('Matches')
plt.legend();
````
